```python
# Create a dictionary with keys 1â€“5 and values as their squares
```

```python
d = {i: i*i for i in range(1, 6)}
print(d)
```

```python
# Access a deeply nested key safely in a dictionary
```

```python
d = {"a": {"b": {"c": 50}}}
print(d.get("a", {}).get("b", {}).get("c"))
```

```python
# Invert a dictionary (swap keys and values)
```

```python
d = {"a": 1, "b": 2, "c": 3}
inv = {v: k for k, v in d.items()}
print(inv)
```

```python
# Merge two dictionaries, summing values of common keys
```

```python
d1 = {"a": 5, "b": 3}
d2 = {"b": 7, "c": 2}
result = {k: d1.get(k, 0) + d2.get(k, 0) for k in set(d1) | set(d2)}
print(result)
```

```python
# Check if two dictionaries are identical
```

```python
d1 = {"a":1, "b":2}
d2 = {"b":2, "a":1}
print(d1 == d2)
```

```python
# Convert dictionary to a list of "key=value" strings
```

```python
d = {"a":10, "b":20}
result = [f"{k}={v}" for k,v in d.items()]
print(result)
```
