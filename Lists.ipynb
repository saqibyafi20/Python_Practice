```python
# Find the Smallest Element in a List.
```

```python
nums = [10, 25, 3, 47, 19]
smallest = (min(nums))
print(smallest)
```

```python
# Reverse a List
```

```python
nums = [1, 2, 3, 4]
nums.reverse()
print(nums)
```

```python
# Remove duplicates from a list while preserving the original order.
```

```python
nums = [1, 2, 3, 2, 4, 1, 5]
unique = []

for n in nums:
    if n not in unique:
        unique.append(n)

print(unique)
```

```python
# Find the second largest unique number in a list.
```

```python
nums = [12, 45, 2, 45, 33, 33, 44]

unique_nums = list(set(nums))
unique_nums.sort(reverse=True)

print("Second largest:", unique_nums[1])
```

```python
# Find Common Elements Between Two Lists.
```

```python
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

common = [x for x in a if x in b]
print(common)
```

```python
# Multiply All Items in a List.
```

```python
nums = [2, 3, 4]
product = 1
for n in nums:
    product *= n

print("Product:", product)
```
