```python
# Extract every 3rd element from a list
```

```python
nums = [10, 20, 30, 40, 50, 60, 70, 80, 90]
result = nums[2::3]
print(result)
```

```python
# Reverse a list using slicing
```

```python
items = ["a", "b", "c", "d"]
rev = items[::-1]
print(rev)
```

```python
# Get the middle half of a list
```

```python
data = [1,2,3,4,5,6,7,8]
n = len(data)
mid = data[n//4 : 3*n//4]
print(mid)
```

```python
# Remove the first and last 2 elements using slicing
```

```python
lst = [5, 10, 15, 20, 25, 30, 35]
trimmed = lst[2:-2]
print(trimmed)
```

```python
# Extract digits at odd positions in a number
```

```python
n = 987654321
s = str(n)
result = s[1::2]
print(result)
```
