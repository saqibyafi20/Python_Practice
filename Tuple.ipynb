```python
# Swap two variables using tuples
```

```python
a, b = 5, 10
a, b = b, a
print(a, b)

```

```python
# Convert a tuple to a comma-separated string
```

```python
t = ("Python", "Java", "C++")
result = ",".join(t)
print(result)
```

```python
# Check if an element exists in a tuple
```

```python
t = (5, 10, 15, 20)
print(15 in t)
```

```python
# Find index of an element in a tuple
```

```python
t = ("a", "b", "c", "d")
print(t.index("c"))
```

```python
# Count occurrences in a tuple
```

```python
t = (1,2,2,3,3,3)
print(t.count(3))
```

```python
# Multiply a tuple
```

```python
t = (1, 2)
print(t * 3)
```
