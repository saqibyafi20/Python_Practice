```python
# Convert List of Strings to Integers.
```

```python
nums = ["1", "2", "3", "4"]
converted = [int(n) for n in nums]
converted
```

```python
# Convert Tuple to List.
```

```python
t = (1, 2, 3)
lst = list(t)
lst
```

```python
# Convert List of Mixed Numbers (int + float) to All Floats.
```

```python
nums = [1, 2.5, 3, 4.75]
floats = [float(x) for x in nums]
floats
```
