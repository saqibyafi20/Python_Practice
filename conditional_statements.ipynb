```python
# A store gives discounts based on purchase amount:

# Above ₹5000 → 20%
# ₹3000-₹5000 → 15%
# Below ₹3000 → 10%
# Additionally:
# If the customer is a member, add extra 5% discount.
# Write a program to compute the final amount after all applicable discounts.
```

```python
amount = 5200
is_member = True
if amount > 5000:
    discount = 0.20
elif 3000 <= amount <= 5000:
    discount = 0.15
else:
    discount = 0.10
if is_member:
    discount += 0.05
final_amount = amount - (amount * discount)
print("Final Amount:", final_amount)
```

```python
# Given marks in three subjects, a student passes only if:

# Average ≥ 60
# No subject has marks < 40
# If average ≥ 90 → Print “Grade A+”
# If 75 ≤ average < 90 → Print “Grade A”
# If 60 ≤ average < 75 → Print “Grade B”
# Otherwise print “Fail”.
```

```python
m1, m2, m3 = 92, 86, 88
avg = (m1 + m2 + m3) / 3
if m1 >= 40 and m2 >= 40 and m3 >= 40:
    if avg >= 90:
        grade = "Grade A+"
    elif avg >= 75:
        grade = "Grade A"
    elif avg >= 60:
        grade = "Grade B"
    else:
        grade = "Fail"
else:
    grade = "Fail"

print("Result:", grade)
```

```python
# Check if a password is strong. Conditions:

# Length ≥ 8
# Contains uppercase
# Contains lowercase
# Contains digit
# Contains special character @#$%^&+=
# Print whether it is strong or weak.
```

```python
password = "Saqib@123"

has_upper = any(c.isupper() for c in password)
has_lower = any(c.islower() for c in password)
has_digit = any(c.isdigit() for c in password)
has_special = any(c in "@#$%^&+=" for c in password)

if len(password) >= 8 and has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

```python
# Given sides a, b, c:
# Check if triangle is valid.
# If valid:
# All sides equal → Equilateral
# Two sides equal → Isosceles
# All sides different → Scalene
```

```python
a, b, c = 5, 5, 8

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Equilateral Triangle")
    elif a == b or a == c or b == c:
        print("Isosceles Triangle")
    else:
        print("Scalene Triangle")
else:
    print("Invalid Triangle")
```

```python
# A year is a leap year if:
# Divisible by 400, or
# Divisible by 4 but not by 100
# Additionally:
# If it is a century year (ends with 00), print “Century Year”.
```

```python
year = 2000

if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")

if year % 100 == 0:
    print("Century Year")
```

```python
# ATM Rules:
# Daily limit: ₹10,000
# ATM balance available: ₹12,000
# Withdraw amount: check
# Must be multiple of ₹100
# Must not exceed daily limit
# Must not exceed ATM balance
```

```python
withdraw = 4500
daily_limit = 10000
atm_balance = 12000

if withdraw % 100 != 0:
    print("Amount must be multiple of 100")
elif withdraw > daily_limit:
    print("Exceeds daily limit")
elif withdraw > atm_balance:
    print("ATM has insufficient cash")
else:
    print("Transaction Successful")
```
