```python
# Write a function to read a file. If the file does not exist, create it and write “New file created”.
```

```python
def read_file(filename):
    try:
        f = open(filename, "r")
        return f.read()
    except FileNotFoundError:
        f = open(filename, "w")
        f.write("New file created")
        return "File not found, so created a new file."

print(read_file("data.txt"))
```

```python
# Use try-except-else-finally to demonstrate all exception blocks in a single program.
```

```python
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("Invalid input!")
else:
    print("You entered:", num)
finally:
    print("Program ended.")
```

```python
# Write a program to raise a custom exception InvalidPasswordError if the password length < 6.
```

```python
class InvalidPasswordError(Exception):
    pass

def validate_password(pwd):
    try:
        if len(pwd) < 6:
            raise InvalidPasswordError("Password too short")
        print("Password accepted")
    except InvalidPasswordError as e:
        print(e)

validate_password("abc")


```

```python
# Write a program that handles division but raises a custom exception if the denominator is zero.
```

```python
class ZeroDenominatorError(Exception):
    pass

def divide(a, b):
    try:
        if b == 0:
            raise ZeroDenominatorError("Denominator cannot be zero!")
        return a / b
    except ZeroDenominatorError as e:
        return e

print(divide(10, 0))
```
