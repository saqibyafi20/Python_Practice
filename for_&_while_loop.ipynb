```python
# Write a program that keeps reading integers from the user until they enter 0.
# Print the sum of all numbers entered.
```

```python
total = 0
num = int(input("Enter number (0 to stop): "))

while num != 0:
    total += num
    num = int(input("Enter number (0 to stop): "))

print("Total:", total)
```

```python
# Given a list of numbers, print only the prime numbers using a while loop.
```

```python
nums = [3, 10, 5, 21, 7, 19]
i = 0

while i < len(nums):
    n = nums[i]
    j = 2
    is_prime = True

    while j * n <= n**2 and j < n:
        if n % j == 0:
            is_prime = False
            break
        j += 1

    if n > 1 and is_prime:
        print(n, "is prime")

    i += 1
```
