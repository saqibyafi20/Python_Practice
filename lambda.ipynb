```python
# Use a lambda to square a number.
```

```python
square = lambda x: x * x
print(square(6))
```

```python
# Use a lambda with map() to convert a list of numbers to their cubes.
```

```python
nums = [1, 2, 3, 4]
cubes = list(map(lambda x: x**3, nums))
print(cubes)
```

```python
# Use a lambda with filter() to keep only even numbers.
```

```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
# Use a lambda to return the last character of a string.
```

```python
last_char = lambda s: s[-1]
print(last_char("lambda"))
```

```python
# Use a lambda to conditionally label numbers as “Even” or “Odd”.
```

```python
label = lambda x: "Even" if x % 2 == 0 else "Odd"
print(label(7))
print(label(10))
```

```python
# Create a lambda that returns the maximum of three numbers.
```

```python
max_three = lambda a, b, c: max(a, b, c)
print(max_three(3, 12, 7))
```
