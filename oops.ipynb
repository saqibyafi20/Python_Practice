```python
# Create a class Person with attributes name and age, and a method introduce()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old."

p = Person("yafu", 25)
print(p.introduce())
```

```python
# Create a class Rectangle with methods to compute area and perimeter.
```

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

rect = Rectangle(5, 3)
print(rect.area())
print(rect.perimeter())
```

```python
# reate a class with private attributes using name mangling.
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private attribute

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
print(acc.get_balance())
```

```python
# Implement encapsulation + getters/setters using property().
```

```python
class Student:
    def __init__(self, grade):
        self._grade = grade

    @property
    def grade(self):
        return self._grade

    @grade.setter
    def grade(self, value):
        if 0 <= value <= 100:
            self._grade = value
        else:
            raise ValueError("Invalid grade")

s = Student(90)
s.grade = 95
print(s.grade)
```
