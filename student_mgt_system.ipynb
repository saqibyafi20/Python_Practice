```python
students = []  # list to store all students

n = int(input("Enter number of students: "))

for i in range(n):
    print(f"\nEnter details for Student {i+1}:")
    name = input("Enter student name: ")
    eng = int(input("Enter marks in English: "))
    math = int(input("Enter marks in Maths: "))
    sci = int(input("Enter marks in Science: "))

    total = eng + math + sci  # arithmetic operators
    avg = total / 3           # average calculation

    if avg >=90:
        grade = "A"
    elif avg >=75:
        grade = "B"
    elif avg >=50:
        grade = "C"
    else:
        grade = "F"


    # store student details in dictionary
    student = {
        "Name": name,
        "English": eng,
        "Maths": math,
        "Science": sci,
        "Total": total,
        "Average": avg,
        "Grade": grade

    }
    students.append(student)

    # add dictionary to list


# Display Final Report
print("\n========== Final Report ==========")
print("Name\tEnglish\tMaths\tScience\tTotal\tAverage\tGrade")
for s in students:
    print(f"{s['Name']}\t{s['English']}\t{s['Maths']}\t{s['Science']}\t{s['Total']}\t{s['Average']:.2f}\t{s['Grade']}")
```
